*Brown corpus is helpful for finding systematic differences between genres*

```python
from nltk.corpus import brown
news_text = brown.words(categories = 'news')
fdist = nltk.FreqDist([w.lower() for w in news_text])
modals = ['can', 'could', 'may', 'might', 'must', 'will']
for m in modals:
    print(m + ':', fdist[m])
```

```python
romance_text = brown.words(categories = 'romance')
fdist = nltk.FreqDist([w.lower() for w in romance_text])
modals = ['who', 'what', 'why', 'how', 'which', 'whom', 'whose', 'love']
for m in modals:
    print(m + ':', fdist[m])
```

```python
# counts for each genre of interest
cfd = nltk.ConditionalFreqDist((genre,word)
    for genre in brown.categories()
    for word in brown.words(categories = 'news'))
genres = ["news", "religion", "hobbies", "science_fiction", "romance", "humor"]
modals = ["can", "could", "may", "might", "must", "will"]
cfd.tabulate(conditions= genres, samples = modals)
```

```python
def plural(word):
    if word.endswith('y'):
        return word[:-1] + 'ies'
    elif word[-1] in 'sx' or word[-2] in ['sh', 'ch']:
        return word + 'es'
    elif word.endswith('an'):
        return word[:-2] + 'en'
    else:
        return word + 's'
print(plural('fairy'))
print(plural('woman'))
print(plural('fish'))   
```

```python
def unsual_words(text):
    text_vocab = set(w.lower() for w in text if w.isalpha())
    english_vocab = set(w.lower() for w in nltk.corpus.words.words())
    unsual = text_vocab.difference(english_vocab)
    return sorted(unsual)
unsual_words(nltk.corpus.gutenberg.words('austen-sense.txt'))
```

```python
unsual_words(nltk.corpus.nps_chat.words())
```

```python
# Stopwords
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
def content_fraction(text):
    stopwords = nltk.corpus.stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords]
    return len(content)/ len(text)
content_fraction(nltk.corpus.reuters.words())
```

*A word puzzle: A grid of randomly chosen letters with rules for creating words out of the
letters; this puzzle is known as Target.*


```python
puzzle_letters = nltk.FreqDist('egivrvonl')
obligatory = 'r'
wordlist = nltk.corpus.words.words()
[w for w in wordlist if len(w) >= 6
    and obligatory in w 
    and nltk.FreqDist(w)<= puzzle_letters]
```

```python
syllable =  ['N', 'IH0', 'K', 'S']
[word for word, pron in entries if pron[-4:]== syllable]
```

```python
sorted(set(w[:2] for w, pron in entries if pron[0] == 'N' and w[0] != 'n'))
```

```python
def stress(pron):
    return [char for phone in pron for char in phone if char.isdigit()]
[w for w, pron in entries if stress(pron) == ['0', '1', '0', '2', '0']]
```

```python
[w for w, pron in entries if stress(pron)== ['0', '2', '0', '1', '0']]
```

```python
p3 = [(pron[0] + '-' + pron[2], word)
     for (word, porn) in entries
     if pron[0] == 'P' and len (pron) == 3]
cfd = nltk.ConditionalFreqDist(p3)
for template in cfd.conditions():
    if len(cfd[template]) > 10:
        words = cfd[template].keys()
        wordlist = ' '.join(words)
        print(template, wordlist[:70] + ".....")
```
